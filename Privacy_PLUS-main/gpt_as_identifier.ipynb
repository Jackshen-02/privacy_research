```python
from collections import defaultdict

# Initialize a defaultdict to count occurrences
word_counts = defaultdict(int)

# Iterate through each document in the DataFrame
for i in range(len(df)):
    tokens = df.iloc[i].tokens  # Access the tokens for the current document
    
    # Update word counts for each token
    for token in tokens:
        word_counts[token] += 1

# Convert defaultdict to a regular dictionary (optional)
word_counts = dict(word_counts)

# Sort the dictionary by occurrences in descending order
sorted_word_counts = sorted(word_counts.items(), key=lambda item: item[1], reverse=True)

# Print the sorted word counts nicely
print("Word Occurrences Summary (sorted by frequency):")
for word, count in sorted_word_counts:
    print(f"{word}: {count}")

# Optionally, save the sorted word counts to a file
# output_file = "sorted_word_occurrences_summary.txt"
# with open(output_file, 'w') as f:
#     for word, count in sorted_word_counts:
#         f.write(f"{word}: {count}\n")

# print(f"Sorted word occurrences summary saved to {output_file}")
```
